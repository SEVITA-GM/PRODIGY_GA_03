```python
import random

def generate_markov_chain(text, chain_length=2):
    words = text.split()
    markov_chain = {}

    for i in range(len(words) - chain_length):
        key = tuple(words[i:i + chain_length])
        next_word = words[i + chain_length]

        if key not in markov_chain:
            markov_chain[key] = []
        markov_chain[key].append(next_word)

    return markov_chain

def generate_text(markov_chain, start_words, length=50):
    words = list(start_words)
    for _ in range(length):
        key = tuple(words[-2:])
        if key in markov_chain:
            next_word = random.choice(markov_chain[key])
            words.append(next_word)
        else:
            break
    return ' '.join(words)

text = """Machine learning is a branch of artificial intelligence that focuses on building systems that can learn from and make decisions based on data.
It enables computers to recognize patterns, make predictions, and improve their performance over time without being explicitly programmed for each task.
Machine learning is widely used in various fields, including healthcare, finance, and marketing, to analyze large datasets, detect anomalies, and provide insights."""
chain_length = 2
markov_chain = generate_markov_chain(text, chain_length)

start_words = ("Machine", "learning")

generated_text = generate_text(markov_chain, start_words)
print("Generated Text:\n", generated_text)
```
